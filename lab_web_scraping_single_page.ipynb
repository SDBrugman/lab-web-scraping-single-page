# Lab web scraping single page

```python
# import libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.billboard.com/charts/hot-100"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
```

```python
songs = soup.findAll('span',attrs={"class":"chart-element__information"})
for x in songs:
    print('songs:', x.find('span', attrs={'class': 'chart-element__information__song text--truncate color--primary'}).text, '| artists:', x.find('span', attrs={'class': 'chart-element__information__artist text--truncate color--secondary'}).text)
```

```python
song_title = []
artist_name = []
year = []
songs = soup.findAll('span',attrs={"class":"chart-element__information"})
for x in songs:
    song = str(x.find('span', attrs={'class': 'chart-element__information__song text--truncate color--primary'}).text).strip()
    artist = str(x.find('span', attrs={'class': 'chart-element__information__artist text--truncate color--secondary'}).text).strip()
    song_title.append(song)
    artist_name.append(artist)
    year.append('2021')
    
df = pd.DataFrame({'song':song_title, 'artist':artist_name, 'year': year})
```

```python
df.head()
```

```python
df.shape
```

```python
#import re
#regex = re.compile('.*__rank.*')
ranks_2021 = []
ranking_2021 = soup.findAll("span", {"class" : "chart-element__rank__number"})
for each in ranking_2021: 
    rankings = str(each.get_text().replace("\n",""))
    ranks_2021.append(rankings)
    #ranks_2021.append(rankings[0])
    
df_rank_2021 = pd.DataFrame({'rank':ranks_2021})
df_rank_2021
```

```python
df_2021 = pd.concat([df,df_rank_2021], axis=1)
df_2021
```

# Expand dataframe

```python
url2 = "https://www.billboard.com/charts/year-end/hot-100-songs"
response2 = requests.get(url2)
soup2 = BeautifulSoup(response2.content, "html.parser")
```

```python
song_title2 = []
artist_name2 = []
year2 = []
songs2 = soup2.findAll('div',attrs={"class":"ye-chart-item__text"})
for x in songs2:
    song = str(x.find('div', attrs={'class': 'ye-chart-item__title'}).text).strip()
    artist = str(x.find('div', attrs={'class': 'ye-chart-item__artist'}).text).strip()
    song_title2.append(song)
    artist_name2.append(artist)
    year2.append('2020')
    
df2 = pd.DataFrame({'song':song_title2, 'artist':artist_name2, 'year': year2})
```

```python
df2
```

```python
import re
regex = re.compile('.*__rank.*')
ranks = []
ranking = soup2.findAll("div", {"class" : regex})
for each in ranking: 
    rankings = str(each.get_text().replace("\n",""))
    ranks.append(rankings)
    
df_rank = pd.DataFrame({'rank':ranks})
df_rank
```

```python
df_rank.shape
```

```python
pd.set_option('display.max_rows', 200)
```

```python
df_2020 = pd.concat([df2,df_rank], axis=1)
df_2020
```

```python
df_all = pd.concat([df_2021,df_2020], ignore_index=True)
df_all
```
